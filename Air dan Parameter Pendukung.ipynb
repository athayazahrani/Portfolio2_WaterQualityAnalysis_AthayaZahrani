## Korelasi Antara Logam Berat di Air dan Parameter Pendukung

```python
df.columns
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting relevant columns for the correlation matrix
columns_of_interest = ['Cd', 'Cr', 'Cu', 'Pb', 'pH', 'Eh', 'Suhu ', 'DO']
correlation_matrix = df[columns_of_interest].corr()

# Create a mask to hide the upper triangle of the correlation matrix
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Creating a heatmap with the mask applied
plt.figure(figsize=(8, 6))

# Customizing the heatmap with additional aesthetics
sns.heatmap(correlation_matrix, 
            mask=mask,  # Apply the mask to hide upper triangle
            annot=True, 
            fmt=".2f", 
            cmap='magma', 
            linewidths=1, 
            linecolor='white', 
            cbar_kws={'shrink': 0.8, 'orientation': 'vertical'}, 
            square=True, 
            annot_kws={'fontsize': 10, 'fontweight': 'bold'})

# Enhancing the plot
plt.title('Korelasi Antara Logam Berat dan Parameter Pendukung', fontsize=14, weight='bold', pad=20)
plt.xticks(fontsize=10, weight='bold', rotation=45)
plt.yticks(fontsize=10, weight='bold')

# Display the heatmap
plt.show()
```

Analisis ini adalah mengenai parameter pendukung (DO, suhu, Eh, dan pH) yang menyebabkan naik atau turunnya konsentrasi logam berat di air. Berdasarkankan hasil visualisasi di atas, korelasi yang pling signifikan adalah antara Cd dan DO (jumlah oksigen yang larut dalam air) sebesar -0.24, yang berarti semakin rendah nilai DO (jumlah oksigen yang terlarut dalam air), maka semakin tinggi konsentrasi Cd.
