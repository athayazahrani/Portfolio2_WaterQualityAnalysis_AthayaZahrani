# **A. *Data Understanding***

**Latar Belakang**

Di Indonesia, pencemaran baik di daratan maupun perairan kerap terjadi, salah satunya adalah pencemaran logam berat yang penting untuk diperhitungkan karena bersifat non-degradable, persisten, dan dapat terakumulasi dalam tubuh makhluk hidup. Logam berat di perairan biasanya terdapat dalam konsentrasi rendah, tetapi aktivitas manusia akan meningkatkan konsentrasinya sehingga menimbulkan masalah lingkungan atau dampak bagi ekosistem. Logam berat yang masuk ke perairan akan terakumulasi dan mengendap dalam sedimen yang pada suatu saat akan terlepas kembali ke kolom air dan menjadi pencemar sekunder di perairan tersebut. Sedimen memegang peranan penting dalam jejak dan akumulasi logam berat yang berpotensi menimbulkan dampak toksisitas terhadap biota air.

Waduk Saguling merupakan waduk pertama dari tiga waduk yang membendung Sungai Citarum Provinsi Jawa Barat. Waduk ini berfungsi sebagai Pembangkit Listrik tenaga Air (PLTA), sumber air untuk irigasi, perikanan jaring terapung, dan sumber air baku untuk kebutuhan domestik dan industri. Berdasarkan hal tersebut, pengetahuan mengenai faktor-faktor yang berkaitan dengan kualitas air seperti kandungan logam berat dalam sedimen menjadi sangat penting. Kualitas air Sungai Citarum Hulu termasuk kategori tercemar berat, bahkan beberapa logam berat terpantau tidak memenuhi baku. Saat ini pemantauan logam berat hanya dilakukan di dalam air sedangkan logam berat memiliki kelarutan yang rendah dalam air, sebagian besar berada dalam sedimen. Berdasarkan hal tersebut, pemantauan konsentrasi logam berat di dalam air saja tidak cukup, diperlukan pemantauan lebih lanjut mengenai logam berat yang mengendap di sedimen.

Dilakukan penelitian laboratorium pada sampel yang diperoleh, `analisis logam berat Cd (Kadmium), Cr (Kromium), Cu (Tembaga), dan Pb (Timbal) yang terkandung di media air (mg/L), sedimen (mg/kg) dan Fraksi F1 (bagian dari logam berat yang lebih mudah larut atau lepas) dilakukan`. `Sampel dimasukkan ke dalam mesin reaktor dan diputar selama 10 menit dengan 2 kecepatan rotasi yang berbeda, yaitu 50 dan 100 Rotations per Minute (RPM). Setelah diputar, sampel diambil per 0, 2, 4, 6, 12, 24, 48, 72, 96, dan 120 menit yang bertujuan untuk melihat kapan kemungkinan logam berat kembali menempel ke sedimen (mengacu pada sifat Deposisi` (Pengendapan kembali partikel dari medium ke permukaan atau dasar). Sampel pada tiap-tiap pengambilan dianalisis dengan harapan: semakin lama sampel didiamkan, konsentrasi logam berat di air semakin kecil, sedangkan di sedimen semakin besar.

```python
from scipy.stats import normaltest

dist = []

for i in range(len(numerical)):
    feature_name = numerical[i]
    stat, p_value = normaltest(df[feature_name].dropna())
    
    if p_value > 0.05:
        distribution = "Normally Distributed"
    else:
        distribution = "Not Normally Distributed"
    
    dist.append([feature_name, stat, p_value, distribution])

dist_df = pd.DataFrame(dist, columns=['Feature', "D'Agostino-Pearson Statistic", 'P-value', 'Distributed'])
dist_df
```

## Distribusi Konsentrasi Logam Berat Berdasarkan Media

## 7. Tren Konsentrasi Logam Berat Berdasarkan Menit Setelah Diputar

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the metals to be plotted
metals = ['Cd', 'Cr', 'Cu', 'Pb']

# Filter data for RPM 50 and 100
filtered_data_50 = df_rpm_50[df_rpm_50['Media'].isin(['Sedimen (mg/kg)', 'F1 (mg/kg)'])]
filtered_data_100 = df_rpm_100[df_rpm_100['Media'].isin(['Sedimen (mg/kg)', 'F1 (mg/kg)'])]

# Creating subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 16))  # 2x2 grid of plots

# Plotting the trend of each metal concentration over time for both RPM 50 and 100
for i, metal in enumerate(metals):
    ax = axes[i//2, i%2]  # Determine the correct subplot location
    
    # Plot for RPM 50
    sns.lineplot(x='Menit Putaran', y=metal, data=filtered_data_50[filtered_data_50['Media'] == 'Sedimen (mg/kg)'], 
                 marker='o', linewidth=2, ax=ax, color='blue', label='Sedimen - RPM 50')
    sns.lineplot(x='Menit Putaran', y=metal, data=filtered_data_50[filtered_data_50['Media'] == 'F1 (mg/kg)'], 
                 marker='o', linewidth=2, ax=ax, color='cyan', label='F1 - RPM 50')
    
    # Plot for RPM 100
    sns.lineplot(x='Menit Putaran', y=metal, data=filtered_data_100[filtered_data_100['Media'] == 'Sedimen (mg/kg)'], 
                 marker='o', linewidth=2, ax=ax, color='orange', label='Sedimen - RPM 100')
    sns.lineplot(x='Menit Putaran', y=metal, data=filtered_data_100[filtered_data_100['Media'] == 'F1 (mg/kg)'], 
                 marker='o', linewidth=2, ax=ax, color='red', label='F1 - RPM 100')

    # Enhancing the plot
    ax.set_title(metal, fontsize=18, weight='bold')
    ax.set_xlabel('Menit Setelah Diputar', fontsize=16)
    ax.set_ylabel(f'Konsentrasi {metal} (mg/kg)', fontsize=16)
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.set_facecolor('#f7f7f7')

# Adding a main title to the figure
fig.suptitle('Tren Konsentrasi Logam Berat di Sedimen dan F1', fontsize=20, weight='bold')

# Enhancing the layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Display the plots
plt.show()
```
