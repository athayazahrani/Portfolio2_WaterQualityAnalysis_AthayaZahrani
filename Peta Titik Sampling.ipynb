Analisis ini diharapkan dapat membantu pengambil keputusan dalam melakukan tindakan pengendalian pencemaran air yang telah terjadi Waduk Saguling. Pengambilan sampel air dan sedimen dilakukan di tiga titik, yaitu: Titik 4 (Muara Cihaur Kampung Maroko) pada koordinat 06º53'13.0" N dan 107º25'54.4" E, Titik 6 (Muara Ciminyak-lokasi perikanan jaring terapung) pada koordinat 06º57'14.6" N dan 107º26'03.8" E, dan Titik 9 (dekat Intake structure) pada koordinat 06º54'54.4" N dan 107º22'26.3 E. Pengambilan sampel dilakukan pada Bulan Januari 2018 mewakili musim hujan.

```python
import folium

# Konversi koordinat dari derajat menit detik ke desimal
def dms_to_dd(degrees, minutes, seconds, direction):
    dd = float(degrees) + float(minutes)/60 + float(seconds)/(60*60)
    if direction == 'S' or direction == 'W':
        dd *= -1
    return dd

# Koordinat Titik 4 (Muara Cihaur Kampung Maroko)
titik_4_lat = dms_to_dd(6, 53, 13.0, 'S')  # 06º53'13.0" N -> negatif karena berada di Selatan
titik_4_lon = dms_to_dd(107, 25, 54.4, 'E') # 107º25'54.4" E

# Koordinat Titik 6 (Muara Ciminyak-lokasi perikanan jaring terapung)
titik_6_lat = dms_to_dd(6, 57, 14.6, 'S')  # 06º57'14.6" N -> negatif karena berada di Selatan
titik_6_lon = dms_to_dd(107, 26, 3.8, 'E') # 107º26'03.8" E

# Koordinat Titik 9 (dekat Intake structure)
titik_9_lat = dms_to_dd(6, 54, 54.4, 'S')  # 06º54'54.4" N -> negatif karena berada di Selatan
titik_9_lon = dms_to_dd(107, 22, 26.3, 'E') # 107º22'26.3" E

# Buat peta awal, pusatkan di titik pertama
peta = folium.Map(location=[titik_4_lat, titik_4_lon], zoom_start=12)

# Tambahkan titik-titik ke peta
folium.Marker(location=[titik_4_lat, titik_4_lon], popup="Titik 4: Muara Cihaur Kampung Maroko").add_to(peta)
folium.Marker(location=[titik_6_lat, titik_6_lon], popup="Titik 6: Muara Ciminyak-lokasi perikanan jaring terapung").add_to(peta)
folium.Marker(location=[titik_9_lat, titik_9_lon], popup="Titik 9: dekat Intake structure").add_to(peta)

# Tampilkan peta
peta
peta.save("peta_titik_sampel.html")
```
